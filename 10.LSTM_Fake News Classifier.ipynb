```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('F:/Natural_Language_porcessing/train.csv')
```

```python
data.head()
```

```python
# We are dropping the nan values in the dataset using the dropna
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
# Spliting the data into independent and dependent dataset
```

```python
x = data.drop(columns='label')
y = data['label']
```

```python
x.shape
```

```python
## Importing the lbraries
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
from tensorflow.keras.layers import Dense,LSTM
```

```python
voc_size = 5000
```

```python
x_copy  = x.copy().
```

```python
x_copy.reset_index(inplace = True)
```

```python
x_copy.head(10) # now the index is correct
```

```python
## Data preprocessing 
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer,PorterStemmer
```

```python
lemmat = WordNetLemmatizer()
```

```python
item = []
for i in range(0,len(x_copy)):
    review = re.sub('[^A-Za-z]',' ',x_copy['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmat.lemmatize(word)for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    item.append(review)
```

```python
item
```

```python
## now we got the text data and we have to make this text data into a one_hot format
```

```python
one_hot_code = [one_hot(word,voc_size) for word in item]
```

```python
one_hot_code
```

```python
# highest lenght of string from the list
length = max(item,key = len)
print('The length Size:' ,length)
```

```python
## We are doing making the padding in each sentences
```

```python
sentence_len = 46
embedded_doc = pad_sequences(one_hot_code,maxlen=sentence_len,padding='pre')
print(embedded_doc)
```

```python
embedded_doc[0]
```

```python
vector_features = 45
```

```python
# creating the model
model = Sequential()
model.add(Embedding(voc_size,vector_features,input_length = sentence_len))
model.add(LSTM(100))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
x.shape,y.shape
```

```python
len(embedded_doc),y.shape
```

```python
## chaning this into an array
x_final = np.array(embedded_doc)
y_final = np.array(y)
```

```python
x_final.shape,y_final.shape
```

```python
### Training the model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_final,y_final,test_size = 0.33,random_state = 0)
```

```python
## model trainig
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
y_pred = model.predict_classes(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
### Adding the dropout layer for avoid the overfitting problem
```

```python
from tensorflow.keras.layers import Dropout
```

```python
embedding_vectors = 45
model = Sequential()
model.add(Embedding(voc_size,embedding_vectors,input_length = sentence_len))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=10,batch_size=64)
```

```python
y_prd = model.predict_classes(x_test)
```

```python
accuracy_score(y_test,y_prd)
```
